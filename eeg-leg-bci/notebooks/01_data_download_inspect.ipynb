# 01 — Download & Inspect Dataset

Downloads OpenNeuro `ds004362` (BCI2000 lower-limb motor imagery) and validates the BIDS structure for the runs used in this project (runs 6, 10, 14 — imagined *both feet* trials).

```python
import sys
sys.path.insert(0, '../src')

from pathlib import Path
from download import download_subjects, LEG_RUNS

DATA_DIR = Path('../data')
N_SUBJECTS = 10

print(f'Target runs for leg imagery: {LEG_RUNS}')
```

## Download a subject subset

Set `N_SUBJECTS` above to control how many subjects to fetch. Full dataset is 109 subjects; 10 is enough for a quick local run.

```python
download_subjects(n_subjects=N_SUBJECTS, output_dir=DATA_DIR)
```

## Validate BIDS layout

```python
import mne_bids

bids_path = mne_bids.BIDSPath(root=DATA_DIR)
print(bids_path.match())
```

## Inspect one run

```python
import mne

sample_edf = next(DATA_DIR.glob('sub-001/**/*R06*.edf'))
raw = mne.io.read_raw_edf(sample_edf, preload=True)
print(raw.info)
raw.plot(duration=10, n_channels=20)
```

## Check event annotations (T0/T1/T2)

```python
events, event_id = mne.events_from_annotations(raw)
print('Event IDs:', event_id)
print('Event counts:', {k: (events[:, -1] == v).sum() for k, v in event_id.items()})
```
