# 06 — Results Visualisation

Combines classical + EEGNet metrics into `results/metrics.json` and generates the figures referenced in the README and dashboard: model comparison bars, confusion matrices, and band-power summaries.

```python
import sys
sys.path.insert(0, '../src')

import json
from pathlib import Path

from evaluate import plot_accuracy_comparison, plot_confusion_matrices, summarize

RESULTS_DIR = Path('../results')

classical = json.loads((RESULTS_DIR / 'classical_metrics.json').read_text())
eegnet = json.loads((RESULTS_DIR / 'eegnet_metrics.json').read_text())
all_results = classical + [eegnet]

with open(RESULTS_DIR / 'metrics.json', 'w') as f:
    json.dump(all_results, f, indent=2)
print('Combined results written to results/metrics.json')
```

## Summary table

```python
summarize(all_results)
```

## Model comparison bar chart

```python
plot_accuracy_comparison(all_results, RESULTS_DIR / 'figures')
```

## Confusion matrices

```python
plot_confusion_matrices(all_results, RESULTS_DIR / 'figures')
```

## Scalp topomap — alpha-band ERD (left vs right foot imagery)

Requires MNE channel positions; uses the `standard_1005` montage set during preprocessing.

```python
import mne
import numpy as np
import matplotlib.pyplot as plt

from utils import CHANNELS

# Placeholder ERD values per channel for illustration —
# replace with real per-channel alpha-band log-power differences
# computed from your epochs (left - right).
rng = np.random.default_rng(42)
erd_values = rng.normal(0, 1, len(CHANNELS))

info = mne.create_info(CHANNELS, sfreq=160, ch_types='eeg')
info.set_montage('standard_1005', on_missing='warn')

fig, ax = plt.subplots(figsize=(4, 4))
mne.viz.plot_topomap(erd_values, info, axes=ax, show=False, cmap='RdBu_r')
ax.set_title('Alpha-band ERD: Left vs Right foot imagery')
plt.show()
```
